```python
from selenium.common.exceptions import NoSuchElementException, StaleElementReferenceException
import unicodedata
def get_review(url):
    driver = webdriver.Chrome('chromedriver.exe')
    driver.get(url)
    time.sleep(2)
    reviews = {}
    shelves = {}
    i=1
    book = BeautifulSoup(requests.get(url).text)
    #print driver.current_url
    title = list(book.find('h1',id = 'bookTitle').strings)[0].strip()
    print title
    while i<20:
        #click and expand all the reviews
        soup = BeautifulSoup(driver.page_source,'html.parser')
        rev_divs = None
        rev_divs = soup.find_all('div',class_ = 'left bodycol')
        #print i
        for rev_div in rev_divs:
            rev = None
            rev = rev_div.find('span', {'id': re.compile('freeText'), 'style': 'display:none'}) 
            ID = rev_div.find('a',class_ = 'user').get('title')
            user = rev_div.find('a',class_ = 'user').get('title')
#             print ID
            try:
            #if rev.find('a',class_ = 'actionLinkLite') != None:
                #print rev_div.find_all('a',class_ = 'actionLinkLite')
                shelf = map(lambda x: x.string, rev_div.find_all('a',class_ = 'actionLinkLite'))
                #print shelf
            except AttributeError:
                #print 'error'
                shelf = None
 
            if not rev:
                rev = rev_div.find('span', {'id':re.compile('freeText')})
            
            if rev:
                rev = rev.text.encode('utf-8')
                 
            reviews[ID]= rev
            shelves[user]= shelf
            
        try:
            i = i+1
            page_num = driver.find_element_by_class_name('next_page').click()
            time.sleep(2)
            print 'page %d' %i
        except NoSuchElementException as e:
            print e
            break
        except StaleElementReferenceException as e:
            print e
            break
    resultdf1 = pd.DataFrame(reviews.items(), columns = ['user', 'review'])
    resultdf1.to_csv(title+'review.csv', encoding='utf-8')
    resultdf2 = pd.DataFrame(shelves.items(),columns = ['user','tags'])
    resultdf2.to_csv(title+'tags.csv', encoding = 'utf-8')
    return resultdf1, resultdf2
```
